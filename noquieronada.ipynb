# **Preguntas**

## ¿Qué municipios tienen más accesos por cada 1000 habitantes en un año y trimestre específicos?


```python
año = 2023
trimestre = 3

df_filtrado = df[(df['AÑO'] == año) & (df['TRIMESTRE'] == trimestre)]

# Calcular índice de accesos por cada 1000 habitantes
df_filtrado['ACCESOS_POR_1000'] = (df_filtrado['No. ACCESOS FIJOS A INTERNET'] / df_filtrado['POBLACIÓN DANE']) * 1000

# Ordenar de mayor a menor
top_municipios = df_filtrado.sort_values('ACCESOS_POR_1000', ascending=False)

# Mostrar los 10 primeros
print(top_municipios[['DEPARTAMENTO', 'MUNICIPIO', 'No. ACCESOS FIJOS A INTERNET', 'POBLACIÓN DANE', 'ACCESOS_POR_1000']].head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Gráfica de barras
plt.figure(figsize=(10, 5))
bars = plt.bar(top_municipios['MUNICIPIO'].head(), top_municipios['ACCESOS_POR_1000'].head(), color='green')
plt.title('Índice de Internet por municipio')
plt.xlabel('Municipio')
plt.ylabel('Índice (%)')
plt.xticks(rotation=45)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## ¿Cómo ha crecido el acceso a internet en un departamento entre trimestres?

```python


# Seleccionar el departamento
departamento = 'CASANARE'

# Filtrar y agrupar por año y trimestre
df_dep = df[df['DEPARTAMENTO'].str.upper() == departamento.upper()].copy()
df_grouped = df_dep.groupby(['AÑO', 'TRIMESTRE'])['No. ACCESOS FIJOS A INTERNET'].sum().reset_index()

# Ordenar por año y trimestre
df_grouped.sort_values(by=['AÑO', 'TRIMESTRE'], inplace=True)

# Crear lista de trimestres (1, 2, 3, 4, ...) y años
x = list(range(len(df_grouped)))
y = df_grouped['No. ACCESOS FIJOS A INTERNET']

# Crear etiquetas para trimestres: T1, T2, T3, T4...
trimestres = ['T' + str(t) for t in df_grouped['TRIMESTRE']]

# Crear etiquetas de año, solo una vez cada 4 trimestres
etiquetas_anio = []
for i, (anio, trimestre) in enumerate(zip(df_grouped['AÑO'], df_grouped['TRIMESTRE'])):
    etiquetas_anio.append(str(anio) if trimestre == 1 else '')

# Graficar
plt.figure(figsize=(14, 6))
plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.xticks(ticks=x, labels=trimestres)
plt.title(f'Evolución del acceso fijo a Internet en {departamento.title()} por trimestre')
plt.xlabel('Trimestres')
plt.ylabel('No. de accesos fijos a Internet')
plt.grid(True, linestyle='--', alpha=0.5)

# Agregar los años debajo de los trimestres (como una segunda línea de etiquetas)
# Usamos anotaciones para simular una segunda línea de texto bajo el eje X


plt.tight_layout()
plt.show()
```
