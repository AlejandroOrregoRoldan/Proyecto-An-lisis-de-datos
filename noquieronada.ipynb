```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.ticker import FuncFormatter

def generar_grafico_evolucion(df, departamento_seleccionado):
    """
    Genera un gráfico de líneas con la evolución de accesos a internet para
    un departamento específico, usando la lógica proporcionada.

    Args:
        df (pd.DataFrame): El DataFrame con todos los datos.
        departamento_seleccionado (str): El nombre del departamento a analizar.

    Returns:
        matplotlib.figure.Figure: El objeto de la figura con el gráfico.
    """
    # --- 2. Lógica de Filtrado y Agrupación ---
    df_dep = df[df['DEPARTAMENTO'].str.upper() == departamento_seleccionado.upper()].copy()
    df_grouped = df_dep.groupby(['AÑO', 'TRIMESTRE'])['No. ACCESOS FIJOS A INTERNET'].sum().reset_index()
    df_grouped.sort_values(by=['AÑO', 'TRIMESTRE'], inplace=True)

    # --- 3. Lógica de Graficación ---
    fig, ax = plt.subplots(figsize=(16, 8))
    x = range(len(df_grouped))
    y = df_grouped['No. ACCESOS FIJOS A INTERNET']
    trimestres_labels = [f"T{t}" for t in df_grouped['TRIMESTRE']]
    ax.plot(x, y, marker='o', linestyle='-', color='#007ACC', linewidth=2, markersize=8, label='No. de Accesos')

    # --- 4. Personalización de Ejes y Títulos ---
    ax.set_title(f'Evolución de Accesos Fijos a Internet en {departamento_seleccionado.title()}', fontsize=18, weight='bold')
    ax.set_xlabel('Año y Trimestre', fontsize=12)
    ax.set_ylabel('Número de Accesos Fijos', fontsize=12)
    ax.set_xticks(x)
    ax.set_xticklabels(trimestres_labels, rotation=45, ha="right")

    # Definir el número de marcas en el eje Y
    num_ticks_y = 12
    y_min, y_max = ax.get_ylim()
    ax.set_yticks(np.linspace(y_min, y_max, num=num_ticks_y))

    # Añadir etiquetas de año
    y_pos_anio = ax.get_ylim()[0]
    for i, row in df_grouped.iterrows():
        if row['TRIMESTRE'] == 1:
            ax.text(i, y_pos_anio, row['AÑO'], ha='center', va='bottom', fontsize=11, weight='bold')

    ax.grid(True, which='major', linestyle='--', linewidth='0.5', color='grey')
    fig.tight_layout()
    
    return fig

# --- Celda para ejecutar en Jupyter ---

# 1. Cargar los datos y preparar el DataFrame
sns.set_theme(style="whitegrid")
df = pd.read_csv('Internet_Fijo_Penetraci_n_Municipio_20250804.csv')
df['No. ACCESOS FIJOS A INTERNET'] = df['No. ACCESOS FIJOS A INTERNET'].astype(int)

# 2. Definir el departamento y llamar a la función
departamento = 'chocó'
figura_chocó = generar_grafico_evolucion(df, departamento)

# 3. Mostrar la gráfica en la salida de la celda
plt.show()

# Opcional: si también quieres guardarla como imagen
# figura_chocó.savefig('evolucion_choco.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Carga y Preparación de Datos ---
sns.set_theme(style="whitegrid")
df = pd.read_csv('Internet_Fijo_Penetraci_n_Municipio_20250804.csv')
df['No. ACCESOS FIJOS A INTERNET'] = df['No. ACCESOS FIJOS A INTERNET'].astype(int)

# --- 2. Lógica de Cálculo por Departamento (Tu Lógica Adaptada) ---

# Definir el periodo de análisis
año = 2023
trimestre = 3

# Filtrar el DataFrame para el año y trimestre específicos
df_filtrado = df[(df['AÑO'] == año) & (df['TRIMESTRE'] == trimestre)]

# Agrupar por departamento, sumando los accesos y la población
df_agregado = df_filtrado.groupby('DEPARTAMENTO').agg({
    'No. ACCESOS FIJOS A INTERNET': 'sum',
    'POBLACIÓN DANE': 'sum'
}).reset_index()

# Calcular el índice de penetración por cada 100 habitantes para cada departamento
df_agregado['INDICE_CALCULADO'] = (df_agregado['No. ACCESOS FIJOS A INTERNET'] / df_agregado['POBLACIÓN DANE']) * 100

# Ordenar para obtener el Top 10
top_10_deptos = df_agregado.sort_values('INDICE_CALCULADO', ascending=False).head(10)


# --- 3. Visualización con el Diseño Solicitado ---

# Crear la figura con el tamaño deseado
plt.figure(figsize=(12, 8))

# Usar seaborn para crear la gráfica de barras con la paleta 'viridis'
sns.barplot(
    x=top_10_deptos['INDICE_CALCULADO'], 
    y=top_10_deptos['DEPARTAMENTO'], 
    palette='viridis'
)

# Personalizar títulos y etiquetas con el estilo que te gustó
plt.title(f'Top 10 Departamentos por Penetración de Internet ({año}-T{trimestre})', fontsize=16, weight='bold')
plt.xlabel('Accesos por cada 100 Habitantes (%)', fontsize=12)
plt.ylabel('Departamento', fontsize=12)

# Ajustar layout y mostrar la gráfica
plt.tight_layout()
plt.show()
```

```python
import tablero as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Carga de Datos (Cacheada para mayor eficiencia) ---
@st.cache_data
def cargar_datos():
    """Carga y prepara el DataFrame una sola vez."""
    df = pd.read_csv('Internet_Fijo_Penetraci_n_Municipio_20250804.csv')
    df['No. ACCESOS FIJOS A INTERNET'] = df['No. ACCESOS FIJOS A INTERNET'].astype(int)
    return df

# --- 2. Función Principal y Flexible para Generar Gráficos ---
def crear_grafico_ranking(df, año, trimestre, nivel_geografico, tipo_ranking, n=10):
    """
    Genera un gráfico de ranking de penetración de internet con múltiples opciones.

    Args:
        df (pd.DataFrame): El DataFrame principal.
        año (int): Año a analizar.
        trimestre (int): Trimestre a analizar.
        nivel_geografico (str): 'DEPARTAMENTO' o 'MUNICIPIO'.
        tipo_ranking (str): 'mejores' o 'peores'.
        n (int): El número de resultados a mostrar.

    Returns:
        tuple: (matplotlib.figure.Figure, pd.DataFrame)
    """
    # Filtrar por periodo
    df_periodo = df[(df['AÑO'] == año) & (df['TRIMESTRE'] == trimestre)].copy()

    # Procesar según el nivel geográfico
    if nivel_geografico == 'DEPARTAMENTO':
        data_proc = df_periodo.groupby('DEPARTAMENTO').agg({
            'No. ACCESOS FIJOS A INTERNET': 'sum',
            'POBLACIÓN DANE': 'sum'
        }).reset_index()
    else: # Para 'MUNICIPIO'
        data_proc = df_periodo.copy()

    # Calcular el índice, evitando división por cero
    data_proc = data_proc[data_proc['POBLACIÓN DANE'] > 0]
    data_proc['INDICE_CALCULADO'] = (data_proc['No. ACCESOS FIJOS A INTERNET'] / data_proc['POBLACIÓN DANE']) * 100

    # Ordenar según el tipo de ranking
    ascendente = (tipo_ranking == 'peores')
    if ascendente:
        # Para 'peores', filtramos los que tienen conectividad para un ranking más útil
        data_proc = data_proc[data_proc['No. ACCESOS FIJOS A INTERNET'] > 0]
    
    data_ranked = data_proc.sort_values('INDICE_CALCULADO', ascending=ascendente).head(n)

    # Si no hay datos después de filtrar, retornar vacío
    if data_ranked.empty:
        fig, ax = plt.subplots()
        ax.text(0.5, 0.5, "No hay datos para los criterios seleccionados", ha='center')
        return fig, pd.DataFrame()

    # Visualización
    titulo_ranking = "Mejores" if tipo_ranking == 'mejores' else "Peores"
    titulo_geo = "Departamentos" if nivel_geografico == 'DEPARTAMENTO' else "Municipios"
    
    fig, ax = plt.subplots(figsize=(12, 8))
    sns.barplot(
        x=data_ranked['INDICE_CALCULADO'],
        y=data_ranked[nivel_geografico],
        palette='viridis' if tipo_ranking == 'mejores' else 'rocket_r',
        ax=ax
    )
    
    ax.set_title(f'{titulo_ranking} {n} {titulo_geo} por Penetración de Internet ({año}-T{trimestre})', fontsize=16, weight='bold')
    ax.set_xlabel('Accesos por cada 100 Habitantes (%)', fontsize=12)
    ax.set_ylabel(titulo_geo[:-1] if nivel_geografico.endswith('s') else titulo_geo, fontsize=12)
    fig.tight_layout()

    return fig, data_ranked

# --- 3. Construcción de la Interfaz de Streamlit ---

# Cargar los datos al iniciar la app
df_main = cargar_datos()

st.title("📊 Visualizador Interactivo de Penetración de Internet")
st.markdown("Usa los filtros para explorar la conectividad en Colombia a diferentes niveles.")

# Crear columnas para los controles
st.sidebar.header("Filtros de Visualización")
año_sel = st.sidebar.selectbox("Año", sorted(df_main['AÑO'].unique(), reverse=True))
trim_sel = st.sidebar.selectbox("Trimestre", sorted(df_main['TRIMESTRE'].unique()))
geo_sel = st.sidebar.radio("Nivel Geográfico", ["DEPARTAMENTO", "MUNICIPIO"])
ranking_sel = st.sidebar.radio("Tipo de Ranking", ["Mejores 10", "Peores 10"])

# Mapear la selección del ranking a un valor para la función
tipo_ranking_map = 'mejores' if ranking_sel == "Mejores 10" else 'peores'

# --- 4. Generación y Visualización de Resultados ---

st.header(f"Resultados para: {ranking_sel} de {geo_sel}s")

# Llamar a la función principal con las selecciones del usuario
figura, datos = crear_grafico_ranking(
    df=df_main,
    año=año_sel,
    trimestre=trim_sel,
    nivel_geografico=geo_sel,
    tipo_ranking=tipo_ranking_map,
    n=10
)

# Mostrar la gráfica y los datos
if not datos.empty:
    st.pyplot(figura)
    
    with st.expander("Ver tabla de datos"):
        st.dataframe(datos)
else:
    st.warning("No se encontraron datos para la combinación de filtros seleccionada.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Definición de las Funciones ---

def crear_grafico_ranking(df, año, trimestre, nivel_geografico='DEPARTAMENTO', tipo_ranking='mejores', n=10):
    """
    Función principal y flexible para generar gráficos de ranking.
    """
    df_periodo = df[(df['AÑO'] == año) & (df['TRIMESTRE'] == trimestre)].copy()
    if nivel_geografico == 'DEPARTAMENTO':
        data_proc = df_periodo.groupby('DEPARTAMENTO').agg({
            'No. ACCESOS FIJOS A INTERNET': 'sum',
            'POBLACIÓN DANE': 'sum'
        }).reset_index()
    else:
        data_proc = df_periodo.copy()
    
    data_proc = data_proc[data_proc['POBLACIÓN DANE'] > 0]
    data_proc['INDICE_CALCULADO'] = (data_proc['No. ACCESOS FIJOS A INTERNET'] / data_proc['POBLACIÓN DANE']) * 100
    
    ascendente = (tipo_ranking == 'peores')
    if ascendente:
        data_proc = data_proc[data_proc['No. ACCESOS FIJOS A INTERNET'] > 0]
    
    data_ranked = data_proc.sort_values('INDICE_CALCULADO', ascending=ascendente).head(n)

    titulo_ranking = "Mejores" if tipo_ranking == 'mejores' else "Peores"
    titulo_geo = "Departamentos" if nivel_geografico == 'DEPARTAMENTO' else "Municipios"
    
    fig, ax = plt.subplots(figsize=(12, 8))
    sns.barplot(x=data_ranked['INDICE_CALCULADO'], y=data_ranked[nivel_geografico],
                palette='viridis' if tipo_ranking == 'mejores' else 'rocket_r', ax=ax)
    
    ax.set_title(f'{titulo_ranking} {n} {titulo_geo} por Penetración ({año}-T{trimestre})', fontsize=16, weight='bold')
    ax.set_xlabel('Accesos por cada 100 Habitantes (%)', fontsize=12)
    ax.set_ylabel(titulo_geo[:-1], fontsize=12)
    fig.tight_layout()

    return fig, data_ranked

def obtener_top_10(df, año, trimestre, nivel_geografico):
    return crear_grafico_ranking(df, año, trimestre, nivel_geografico, 'mejores', 10)

def obtener_peores_10(df, año, trimestre, nivel_geografico):
    return crear_grafico_ranking(df, año, trimestre, nivel_geografico, 'peores', 10)


# --- 2. Ejecución Principal del Script ---

if __name__ == "__main__":
    print("Iniciando análisis...")
    
    # Cargar los datos
    try:
        df = pd.read_csv('Internet_Fijo_Penetraci_n_Municipio_20250804.csv')
        df['No. ACCESOS FIJOS A INTERNET'] = df['No. ACCESOS FIJOS A INTERNET'].astype(int)
        print("Datos cargados exitosamente.")
    except FileNotFoundError:
        print("Error: No se encontró el archivo 'Internet_Fijo_Penetraci_n_Municipio_20250804.csv'.")
        exit()

    # Definir los parámetros para el análisis
    año_analisis = 2024
    trimestre_analisis = 2
    nivel_analisis = 'DEPARTAMENTO' # Puedes cambiarlo a 'MUNICIPIO'

    # --- Generar y Guardar Gráfica de los MEJORES ---
    fig_mejores, datos_mejores = obtener_top_10(df, año_analisis, trimestre_analisis, nivel_analisis)
    
    # Guardar la gráfica
    nombre_grafica_mejores = f'mejores_{nivel_analisis.lower()}.png'
    fig_mejores.savefig(nombre_grafica_mejores)
    print(f"-> Gráfica guardada como: {nombre_grafica_mejores}")
    
    # Guardar los datos en un archivo CSV
    nombre_tabla_mejores = f'datos_mejores_{nivel_analisis.lower()}.csv'
    datos_mejores.to_csv(nombre_tabla_mejores, index=False)
    print(f"-> Tabla de datos guardada como: {nombre_tabla_mejores}")
    
    
    # --- Generar y Guardar Gráfica de los PEORES ---
    fig_peores, datos_peores = obtener_peores_10(df, año_analisis, trimestre_analisis, nivel_analisis)

    # Guardar la gráfica
    nombre_grafica_peores = f'peores_{nivel_analisis.lower()}.png'
    fig_peores.savefig(nombre_grafica_peores)
    print(f"-> Gráfica guardada como: {nombre_grafica_peores}")

    # Guardar los datos en un archivo CSV
    nombre_tabla_peores = f'datos_peores_{nivel_analisis.lower()}.csv'
    datos_peores.to_csv(nombre_tabla_peores, index=False)
    print(f"-> Tabla de datos guardada como: {nombre_tabla_peores}")

    print("\nAnálisis finalizado.")
```

## Distribución Porcentual de Accesos a Internet por Departamento

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Definición de la Función ---
def crear_torta_accesos_por_departamento(df, top_n=10):
    # (La función es la misma que te di antes, no necesita cambios)
    accesos_por_depto = df.groupby('DEPARTAMENTO')['No. ACCESOS FIJOS A INTERNET'].sum().sort_values(ascending=False)
    if len(accesos_por_depto) > top_n:
        top_deptos = accesos_por_depto.head(top_n)
        otros_sum = accesos_por_depto.iloc[top_n:].sum()
        data_to_plot = pd.concat([top_deptos, pd.Series({'Otros': otros_sum})])
    else:
        data_to_plot = accesos_por_depto
    
    fig, ax = plt.subplots(figsize=(12, 10))
    colors = sns.color_palette('viridis_r', len(data_to_plot))
    wedges, texts, autotexts = ax.pie(
        data_to_plot, 
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        pctdistance=0.85,
        wedgeprops=dict(width=0.4)
    )
    plt.setp(autotexts, size=10, weight="bold", color="white")
    ax.set_title('Distribución Porcentual de Accesos a Internet por Departamento', fontsize=16, weight='bold')
    ax.legend(wedges, data_to_plot.index,
              title="Departamentos",
              loc="center left",
              bbox_to_anchor=(1, 0, 0.5, 1))
    return fig

# --- 2. Código de Ejecución para un Script (.py) ---

# Cargar los datos
df = pd.read_csv('Internet_Fijo_Penetraci_n_Municipio_20250804.csv')
df['No. ACCESOS FIJOS A INTERNET'] = df['No. ACCESOS FIJOS A INTERNET'].astype(int)

# Llamar a la función para generar la figura
grafico_torta_fig = crear_torta_accesos_por_departamento(df)

# --- INICIO DE LA MODIFICACIÓN ---
# Guardar la figura en un archivo en lugar de solo mostrarla
nombre_archivo = 'grafico_torta_departamentos.png'
grafico_torta_fig.savefig(nombre_archivo, bbox_inches='tight')

print(f"Gráfica guardada exitosamente como: {nombre_archivo}")
# Opcional: Si aún quieres que intente abrir la ventana emergente, descomenta la siguiente línea
# plt.show()
# --- FIN DE LA MODIFICACIÓN ---
```
