# **Preguntas**

## ¿Qué municipios tienen más accesos por cada 1000 habitantes en un año y trimestre específicos?


```python
año = 2023
trimestre = 3

df_filtrado = df[(df['AÑO'] == año) & (df['TRIMESTRE'] == trimestre)]

# Calcular índice de accesos por cada 100 habitantes
df_filtrado['ACCESOS_POR_100'] = (df_filtrado['No. ACCESOS FIJOS A INTERNET'] / df_filtrado['POBLACIÓN DANE']) * 100

# Ordenar de mayor a menor
top_municipios = df_filtrado.sort_values('ACCESOS_POR_100', ascending=False)

# Mostrar los 10 primeros
print(top_municipios[['DEPARTAMENTO', 'MUNICIPIO', 'No. ACCESOS FIJOS A INTERNET', 'POBLACIÓN DANE', 'ACCESOS_POR_100']].head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Gráfica de barras
plt.figure(figsize=(10, 5))
bars = plt.bar(top_municipios['MUNICIPIO'].head(), top_municipios['ACCESOS_POR_100'].head(), color='green')
plt.title('Índice de Internet por municipio')
plt.xlabel('Municipio')
plt.ylabel('Índice (%)')
plt.xticks(rotation=45)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## ¿Cómo ha crecido el acceso a internet en un departamento entre trimestres?

```python


# Seleccionar el departamento
departamento = 'CASANARE'

# Filtrar y agrupar por año y trimestre
df_dep = df[df['DEPARTAMENTO'].str.upper() == departamento.upper()].copy()
df_grouped = df_dep.groupby(['AÑO', 'TRIMESTRE'])['No. ACCESOS FIJOS A INTERNET'].sum().reset_index()

# Ordenar por año y trimestre
df_grouped.sort_values(by=['AÑO', 'TRIMESTRE'], inplace=True)

# Crear lista de trimestres (1, 2, 3, 4, ...) y años
x = list(range(len(df_grouped)))
y = df_grouped['No. ACCESOS FIJOS A INTERNET']

# Crear etiquetas para trimestres: T1, T2, T3, T4...
trimestres = ['T' + str(t) for t in df_grouped['TRIMESTRE']]

# Crear etiquetas de año, solo una vez cada 4 trimestres
etiquetas_anio = []
for i, (anio, trimestre) in enumerate(zip(df_grouped['AÑO'], df_grouped['TRIMESTRE'])):
    etiquetas_anio.append(str(anio) if trimestre == 1 else '')

# Graficar
plt.figure(figsize=(14, 6))
plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.xticks(ticks=x, labels=trimestres)
plt.title(f'Evolución del acceso fijo a Internet en {departamento.title()} por trimestre')
plt.xlabel('Trimestres')
plt.ylabel('No. de accesos fijos a Internet')
plt.grid(True, linestyle='--', alpha=0.5)

# Agregar los años debajo de los trimestres (como una segunda línea de etiquetas)
# Usamos anotaciones para simular una segunda línea de texto bajo el eje X


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.ticker import FuncFormatter # Aunque no la usemos, la dejamos por si la necesitas luego

# --- 1. Carga y Preparación de Datos ---
sns.set_theme(style="whitegrid")
df = pd.read_csv('Internet_Fijo_Penetraci_n_Municipio_20250804.csv')
df['No. ACCESOS FIJOS A INTERNET'] = df['No. ACCESOS FIJOS A INTERNET'].astype(int)

# --- 2. Lógica de Filtrado y Agrupación ---
departamento_seleccionado = 'Antioquia'





df_dep = df[df['DEPARTAMENTO'].str.upper() == departamento_seleccionado.upper()].copy()
df_grouped = df_dep.groupby(['AÑO', 'TRIMESTRE'])['No. ACCESOS FIJOS A INTERNET'].sum().reset_index()
df_grouped.sort_values(by=['AÑO', 'TRIMESTRE'], inplace=True)

# --- 3. Lógica de Graficación ---
fig, ax = plt.subplots(figsize=(16, 8))
x = range(len(df_grouped))
y = df_grouped['No. ACCESOS FIJOS A INTERNET']
trimestres_labels = [f"T{t}" for t in df_grouped['TRIMESTRE']]
ax.plot(x, y, marker='o', linestyle='-', color='#007ACC', linewidth=2, markersize=8, label='No. de Accesos')

# --- 4. Personalización de Ejes y Títulos ---
ax.set_title(f'Evolución de Accesos Fijos a Internet en {departamento_seleccionado.title()}', fontsize=18, weight='bold')
ax.set_xlabel('Año y Trimestre', fontsize=12)
ax.set_ylabel('Número de Accesos Fijos', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(trimestres_labels, rotation=45, ha="right")

num_ticks_y = 12 
y_min, y_max = ax.get_ylim()
ax.set_yticks(np.linspace(y_min, y_max, num=num_ticks_y))

# LA LÍNEA DEL FORMATEADOR HA SIDO ELIMINADA DE AQUÍ

y_pos_anio = ax.get_ylim()[0]
for i, row in df_grouped.iterrows():
    if row['TRIMESTRE'] == 1:
        ax.text(i, y_pos_anio, row['AÑO'], ha='center', va='bottom', fontsize=11, weight='bold')

ax.grid(True, which='major', linestyle='--', linewidth='0.5', color='grey')
fig.tight_layout()
plt.show()
```
